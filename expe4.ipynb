```python
# creation of set
thisset = {"apple","banana","cherry"}
print(thisset)
```

```python
#unordered
thisset = {"apple","banana","cherry"}
print(thisset[1])
```

```python
#duplicates not allowed
thisset = {"apple","banana","cherry","apple"}
print(thisset)
```

```python
# the values True and 1 are considered same value in sets and set does not allow duplicates
thisset = {"apple","banana","cherry",True,1,2}
print(thisset)
```

```python
# to check membership
thisset = {"apple","banana","cherry"}
print("banana" in thisset)
print("grapes" in thisset)
```

```python
# once set is created, you cannot replace its items,
# but you can add new items.
thisset = {"apple","banana","cherry"}
thisset.add("orange")
print(thisset)
```

```python
# remove item
thisset = {"apple","banana","cherry"}
thisset.remove("banana")
print(thisset)
```

```python
# set operations
A = {1,2,3,4}
B = {3,4,5,6}
print("Union: ",A | B)
print("Intersection: ",A & B)
print("Difference1: ",A - B)
print("Difference2: ",B - A)
print("Symmetric Difference: ",A ^ B)
```

```python
#frozenset
x = frozenset({"apple","banana","cherry"})
print(x)
print(type(x))
```

```python
# Unlike sets, elements cannot be added or removed from a frozenset.
x = frozenset({"apple","banana","cherry"})
x.add("orange")
print(x)
```

```python
'''Problem statement 1: Unique Event Participants
A college event has a list of students who registered multiple times by mistake.
Write a program using set to store participants Names and display only Unique
participants.'''
participants = ["a","b","c","d","b","c"]
unique_participants = set(participants)
print(unique_participants)
```

```python
'''Problem statement 2: Three students chose elective subjects.
Find the common subjects chosen by all students using set operation.
Concept used: Set intersection (&)'''
set1 = {"phy","math","chem","history"}
set2 = {"history","math","chem","english"}
set3 = {"phy","math","chem","english"}
print("common subjects: ",set1 & set2 & set3)
```

```python
'''Problem statement 3: One set contains students in cricket club another in football club.
Find students who are:
1)in both clubs
2)in only one club
Concept used: Intersection, Symmetric Difference'''
cricket_club = {"a","b","c","d","e"}
football_club = {"b","d","e","f","g"}
print("in both the clubs: ",cricket_club & football_club)
print("in only one club: ",cricket_club ^ football_club)
```

```python
students = {"a","b","c","d","e"}
present = {"a","c","e"}
print("absent students: ",students - present)
```

```python
'''Problem statement 5: remove invalid entries,a set contains valid codes, remove discontinued course code from set concept: remove() / discard()'''
courses = {"ETC101","ETC102","ETC103"}
courses.discard("ETC102")
print("Updated Courses: ", courses)
```
